## üìå Optimizaci√≥n de Hiperpar√°metros [1.0 puntos]

Para ello, se espera que implementen `Optuna` para optimizar no solo los hiperpar√°metros del modelo, sino tambi√©n los de los preprocesadores utilizados (por ejemplo, OneHot Encoding, Scalers, etc.).

```python
import optuna
import numpy as np
import pandas as pd
from xgboost import XGBClassifier
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, MinMaxScaler
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import f1_score, precision_score, recall_score

# ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
# 0) Sup√≥n que ya tienes:
#    df_train, df_val, drop_cols, preprocessor
# ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî

# 1) Prepara X, y, y calcula balance de clases
X = df_train.drop(columns=drop_cols + ['target'])
y = df_train['target']
X_val = df_val.drop(columns=drop_cols + ['target'])
y_val = df_val['target']

neg = (y == 0).sum()
pos = (y == 1).sum()
ratio = neg / pos
print(f"Ratio negativos/positivos = {ratio:.2f}\n")

# 2) Define la funci√≥n objective para Optuna
def objective(trial):
    # ‚Äî preprocesamiento ‚Äî
    scaler_name = trial.suggest_categorical("scaler", ["none", "standard", "minmax"])
    if scaler_name == "standard":
        scaler = StandardScaler()
    elif scaler_name == "minmax":
        scaler = MinMaxScaler()
    else:
        scaler = "passthrough"

    ohe_drop = trial.suggest_categorical("ohe_drop", [None, "first"])
    ohe = OneHotEncoder(drop=ohe_drop, sparse_output=False, handle_unknown="ignore")

    num_cols = X.select_dtypes(include="number").columns.tolist()
    cat_cols = X.select_dtypes(include="object").columns.tolist()

    pre = ColumnTransformer([
        ("num", scaler, num_cols),
        ("cat", ohe,    cat_cols),
    ])

    # ‚Äî espacio de hiperpar√°metros XGBoost ‚Äî
    params = {
        "n_estimators":       trial.suggest_int("n_estimators",  50, 500),
        "max_depth":          trial.suggest_int("max_depth",     1,  12),
        "learning_rate":      trial.suggest_loguniform("learning_rate", 1e-4, 1.0),
        "subsample":          trial.suggest_float("subsample",     0.5, 1.0),
        "colsample_bytree":   trial.suggest_float("colsample_bytree", 0.5,1.0),
        "gamma":              trial.suggest_float("gamma",      0.0, 5.0),
        "min_child_weight":   trial.suggest_int("min_child_weight", 1, 10),
        "reg_alpha":          trial.suggest_loguniform("reg_alpha",  1e-8, 10.0),
        "reg_lambda":         trial.suggest_loguniform("reg_lambda", 1e-8, 10.0),
        "scale_pos_weight":   trial.suggest_categorical("spw", [1.0, ratio, ratio * 2]),
        "use_label_encoder":  False,
        "eval_metric":        "logloss",
        "random_state":       1,
        "n_jobs":            -1
    }

    clf = XGBClassifier(**params)
    pipe = Pipeline([("pre", pre), ("clf", clf)])

    # ‚Äî CV estratificado con 5 folds ‚Äî
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)
    scores = cross_val_score(pipe, X, y, cv=skf, scoring="f1", n_jobs=-1)
    return scores.mean()


```

```python
# 3) Ejecuta la optimizaci√≥n
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=500)
```

```python
# 4) Inspecciona los rangos realmente probados
records = []
for t in study.trials:
    if t.state.is_finished():
        rec = {"f1": t.value}
        rec.update(t.params)
        records.append(rec)
df_trials = pd.DataFrame(records)
print("\nRangos explorados:")
print(df_trials[["max_depth","learning_rate","gamma","reg_alpha","reg_lambda"]].describe())

# 5) Mejores par√°metros
best_params = study.best_params
print("\n=== Best params ===")
print(best_params)
print(f"Best F1 (CV): {study.best_value:.4f}\n")


# 7) Reconstruye y entrena el pipeline final con best_params
# ‚Äî igual preprocesador que en objective ‚Äî
scaler_name = best_params["scaler"]
scaler_opt  = StandardScaler() if scaler_name=="standard" else MinMaxScaler() if scaler_name=="minmax" else "passthrough"
ohe_opt     = OneHotEncoder(drop=best_params["ohe_drop"], sparse_output=False, handle_unknown="ignore")
preprocessor_opt = ColumnTransformer([
    ("num", scaler_opt, num_cols),
    ("cat", ohe_opt,    cat_cols),
])

clf_opt = XGBClassifier(
    **{k: best_params[k] for k in best_params if k in XGBClassifier().get_params()},
    use_label_encoder=False,
    eval_metric="logloss",
    random_state=1,
    n_jobs=-1
)


full_pipe = Pipeline([("pre", preprocessor_opt), ("clf", clf_opt)])
full_pipe.fit(X, y)
y_pred_opt = full_pipe.predict(X_val)

precision_opt = precision_score(y_val, y_pred_opt)
recall_opt    = recall_score(   y_val, y_pred_opt)
f1_after      = f1_score(       y_val, y_pred_opt)


print("=== Metrics final model ===")
print(f"Precision: {precision_opt:.3f}")
print(f"Recall:    {recall_opt:.3f}")

print(f"F1-Score:                   {f1:.3f}")
print(f"F1 after:  {f1_after:.4f}")
print(f"Improvement: {f1_after - f1:.4f} ({100*(f1_after - f1)/f1:.2f}%)")
```
